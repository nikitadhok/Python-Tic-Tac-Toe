```python
from IPython.display import clear_output

def welcome_to_game():
    
    current_board = ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']
    global game_over
    agreement =''
    
    agreement = str(input('Welcome to Tic Tac Toe Game!! Ready To Play? (Yes or No)'))
    
    while agreement != 'Yes' and agreement != 'No':
        agreement = str(input('Please answer as Yes or No:'))
        if agreement == 'Yes' or agreement == 'No':
            continue
    else:
        if agreement == 'Yes':
            print('  ')
            print('Great! here is your game board:')
            display_board(current_board)
            game_over = 'False'
        elif agreement == 'No':
            print('Thank you! See you soon.')
            game_over = 'True'
```

```python
def display_board(board):

    print('    ')
    print('    ')
    print(board[1],'|',board[2],'|',board[3])
    print('---------')
    print(board[4],'|',board[5],'|',board[6])
    print('---------')
    print(board[7],'|',board[8],'|',board[9])
    print('    ')
```

```python

def player_input():
    
    global preference
    
    preference = input('Who wants to go first ? Player 1(enter X) or Player 2(enter O):')
    
    while preference != 'X' and preference != 'O':
        preference = input('The marker has to be either X or O:')
    return preference
```

```python
def position_input(player):

    global preference
    global positions_used
    
    position = int(input(f'{player} - Choose your position (1-9): '))
  
    while position in positions_used:
        
        position = (int(input('This position already used, enter new position:')))
            
        if position not in range(1,10):
                position = int(input('Choose your position between 1-9: '))
        
    else:
        while position not in range(1,10):
                position = int(input('Choose your position between 1-9: '))
       
    positions_used.append(position)
   
    if player == 'player_1':
        current_board[position] = 'X'
        
    else:
        current_board[position] = 'O'
     
    display_board (current_board)
```

```python
def win_check(board,marker):
    global game_over
    
    winning_dict = {'key1':[1,2,3],'key2':[1,4,7],'key3':[1,5,9],'key4':[2,5,8],'key5':[3,6,9],'key6':[3,5,7],'key7':[4,5,6],'key8':[7,8,9]}
 
    extract_positions = [index for index,value in enumerate(current_board) if value == marker]
    extract_set = set(extract_positions)
        
    if len(extract_set)>=3:
        for v in (winning_dict.values()):
            
            if set(v) <= set(extract_set):
                print(f'Congratulations! {marker} has won the game!')
                game_over = 'True'
                return True
                
```

```python
def if_tie():
    global game_over
    if len(positions_used)== 9:
        print('OOPS!! This is a tie!')
        game_over = 'True'
        return True
```

```python
def if_replay():
    
    return input('Do you want to play again? Enter Yes or No: ').lower().startswith('y')
```

```python

current_board = ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']

def play_game():
    
    while True:
    
        global current_board
        global positions_used
        global player_turn
        positions_used = []
        global preference
        global game_over

        welcome_to_game()
        player_input()
        
        if preference == 'X':
            print('player 1 will go first')
            player_turn = 'player_1'
                
        elif preference == 'O':
            print('player 2 will go first')
            player_turn = 'player_2'
               
        while game_over == 'False':
           
            #FOR PLAYER 1   

            if player_turn == 'player_1':
                position_input('player_1')
                if len(positions_used)>=5:
                    if win_check(current_board,'X'):
                        game_over = 'True'
                    else:
                        if if_tie():
                            break
                        else:
                            player_turn = 'player_2'
                else:
                    player_turn = 'player_2'
                         
             #FOR PLAYER 2        

            if player_turn == 'player_2':
                position_input('player_2')
                if len(positions_used)>=5:
                    if win_check(current_board,'O'):
                        game_over = 'True'
                    else:
                        if if_tie():
                            break
                        else:
                            player_turn = 'player_1'
                else:
                    player_turn = 'player_1'
                
        else:
            if not if_replay():
                break   
```

```python
play_game()
```
